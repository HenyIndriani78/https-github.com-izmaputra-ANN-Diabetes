```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/ANN diabetes/data/diabetes_latih.csv')

X_train = df.values
X_train = np.delete(X_train,8,axis=1)

y_train = df['Outcome'].values

df = pd.read_csv('/content/drive/MyDrive/ANN diabetes/data/diabetes_uji.csv')

X_test = df.values
X_test = np.delete(X_test,8,axis=1)

y_test = df['Outcome'].values

print(X_test)
df.head()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from sklearn.preprocessing import MinMaxScaler

X_train = MinMaxScaler().fit_transform(X_train)
X_test = MinMaxScaler().fit_transform(X_test)

print(X_train)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=3,learning_rate_init=0.1,max_iter=100)

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

round(accuracy_score(y_test,y_pred),3)
```
